```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# result graph 1: the expected size of the kernel

num_vertices_set = [
    100, 133, 178, 237, 316, 422, 562, 745, 
    1000, 1330, 1780, 2370, 3160, 4220, 5620, 7450,
    10000, 13300, 17800, 23700, 31600, 42200, 56200, 74500,
    100000, 133000, 178000, 237000, 316000, 422000, 562000, 745000,
    1000000, 1330000, 1780000, 2370000, 3160000, 4220000, 5620000, 7450000,
    10000000, 
]
alpha_set = ["055", "060", "065", "070", "075", "080", "085", "090"]
alpha_set_label = ["0.55", "0.60", "0.65", "0.70", "0.75", "0.80", "0.85", "0.90"]

result_dir = "../results/reduction_diff_alpha/"

reduct_result = []
for alpha in alpha_set:
    reduct_result.append([])
    for num_vertices in num_vertices_set:
        # file open
        f = open(result_dir + "{}_{}.txt".format(num_vertices, alpha), "r")
        tot = 0
        for i in range(100):
            v, = map(int, f.readline().split())
            tot += v
        reduct_result[-1].append(tot / 100)
        f.close()
```

```python
# plot the result (log-log scale)
markers = ["o", "v", "^", "<", ">", "s", "p", "P", "*", "h", "H", "X", "D", "d", "|", "_"]
plt.figure(figsize=(10, 8))
for i in range(len(alpha_set)):
    plt.plot(num_vertices_set, reduct_result[i], label="alpha={}".format(alpha_set_label[i]), marker=markers[i], linestyle="-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("number of vertices")
plt.ylabel("kernel size")
plt.legend()
plt.xlim(95, 10400000)
plt.grid(True, which="both", ls="-");
plt.show()
```

```python
# result graph 2: the cactus plot of the kernel

num_vertices = 10000000

alpha_set = ["055", "060", "065", "070", "075", "080", "085", "090"]
alpha_set_label = ["0.55", "0.60", "0.65", "0.70", "0.75", "0.80", "0.85", "0.90"]

result_dir = "../results/reduction_diff_alpha/"

kernel_size_set = []

for alpha in alpha_set:
    kernel_size_set.append([0])
    
    # file open
    f = open(result_dir + "{}_{}.txt".format(num_vertices, alpha), "r")
    tot = 0
    for i in range(100):
        v, = map(int, f.readline().split())
        kernel_size_set[-1].append(v)
    kernel_size_set[-1].sort()
    f.close()
```

```python
markers = ["o", "v", "^", "<", ">", "s", "p", "P", "*", "h", "H", "X", "D", "d", "|", "_"]
plt.figure(figsize=(10, 8))
for i in range(len(alpha_set)):
    plt.plot(kernel_size_set[i], label="alpha={}".format(alpha_set_label[i]), marker=markers[i], linestyle="-")
# plt.yscale("log")

x_ticks = np.arange(0, 101, 10)
y_ticks = np.arange(0, 5001, 500)

plt.xticks(x_ticks)
plt.yticks(y_ticks)

plt.xlabel("number of accepted instances")
plt.ylabel("kernel size")
plt.legend()
plt.grid(True, which="both", ls="-")
plt.xlim(1, 100)
# plt.yscale("log")
plt.ylim(0, 5000)
plt.show()
```

```python
# result graph 1: change the C value

num_vertices_set = [
    100, 133, 178, 237, 316, 422, 562, 745, 
    1000, 1330, 1780, 2370, 3160, 4220, 5620, 7450,
    10000, 13300, 17800, 23700, 31600, 42200, 56200, 74500,
    100000, 133000, 178000, 237000, 316000, 422000, 562000, 745000,
    1000000, 
]
alpha = "075"
alpha_label = "0.75"

C_set = ["15", "13", "11", "9", "7"]
C_set_label = ["-15", "-13", "-11", "-9", "-7"]

result_dir = "../results/reduction_diff_C/"

reduct_result = []
for C in C_set:
    reduct_result.append([])
    for num_vertices in num_vertices_set:
        # file open
        f = open(result_dir + "{}_{}_{}.txt".format(num_vertices, alpha, C), "r")
        tot = 0
        for i in range(100):
            v, = map(int, f.readline().split())
            tot += v
        reduct_result[-1].append(tot / 100)
        f.close()
```

```python
# plot the result (log-log scale)
markers = ["o", "v", "^", "<", ">", "s", "p", "P", "*", "h", "H", "X", "D", "d", "|", "_"]
plt.figure(figsize=(8, 6))
for i in range(len(C_set)):
    plt.plot(num_vertices_set, reduct_result[i], label="C={}".format(C_set_label[i]), marker=markers[i], linestyle="-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("number of vertices")
plt.ylabel("kernel size")
plt.legend()
plt.xlim(95, 1040000)
plt.ylim(5, 1000)
plt.grid(True, which="both", ls="-");
plt.show()
```
