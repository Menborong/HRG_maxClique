```python
import networkx as nx
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# version 1 dataset

ver1_x = []
ver1_y = []

DIR_NAME = "./ver1_result/"
graph_size = [1000, 2000, 4000, 8000, 10000]
for n in graph_size:
    FILE_NAME = DIR_NAME + str(n) + "_075.out"
    f = open(FILE_NAME, 'r')

    N, alpha, temp, degree, R = map(float, f.readline().split())
    Q, = map(int, f.readline().split())

    tot_time = 0.0

    for i in range(Q):
        clique_siz, = map(int, f.readline().split())
        clique = list(map(int, f.readline().split()))
        time, = map(float, f.readline().split())
        tot_time += time

    tot_time /= Q
    ver1_x += [n]
    ver1_y += [tot_time]

print(ver1_x, ver1_y)
```
