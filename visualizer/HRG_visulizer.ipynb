```python
import networkx as nx
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
in_dir = "../gen_dataset/dataset/"

f = open(in_dir + '1000_075/1.txt', 'r')
N, M = map(int, f.readline().split())
map(int, f.readline().split())
eds = [list(map(int, f.readline().split())) for _ in range(M)]

f = open(in_dir + '1000_075/1.hyp', 'r')
R, = map(float, f.readline().split())
ps = [list(map(float, f.readline().split())) for _ in range(N)]
```

```python
pos = {i:[ps[i][0]*np.cos(ps[i][1]), ps[i][0]*np.sin(ps[i][1])] for i in range(N)}
```

```python
graph = nx.Graph()
```

```python
graph.add_nodes_from([i for i in range(N)])
graph.add_edges_from(eds)
```

```python
plt.figure(figsize=(12, 12))
cir = plt.Circle((0,0), R, fill=False)
axes = plt.axes(xlim = (-30, 30), ylim = (-30, 30))
axes.add_patch(cir)

nx.draw_networkx_nodes(graph, pos, node_size=2)
nx.draw_networkx_edges(graph, pos, alpha=0.1)
plt.show()
```

```python
f = open('./clique_result.txt', 'r')
NC, = map(int, f.readline().split())
clique = list(map(int, f.readline().split()))
num_red_ed, = map(int, f.readline().split())
red_eds = [list(map(int, f.readline().split())) for _ in range(num_red_ed)]
```

```python
from turtle import color

graph1 = nx.Graph()
graph1.add_nodes_from([i for i in range(N)])
graph1.add_edges_from(eds)

graph2 = nx.Graph()
for a in clique:
    for b in clique:
        if(a >= b): 
            continue
        graph2.add_edge(a, b)

plt.figure(figsize=(12, 12))
cir1 = plt.Circle((0,0), R, fill=False)
cir2 = plt.Circle((0,0), R/2, alpha=0.2)
axes = plt.axes(xlim = (-22, 22), ylim = (-22, 22))
axes.add_patch(cir1)
axes.add_patch(cir2)

nx.draw_networkx_nodes(graph1, pos, node_size=2)
nx.draw_networkx_edges(graph1, pos, alpha=0.1)

nx.draw_networkx_nodes(graph2, pos, node_size=2, node_color='red')
nx.draw_networkx_edges(graph2, pos, alpha=0.9, edge_color='red')
plt.show()
```
